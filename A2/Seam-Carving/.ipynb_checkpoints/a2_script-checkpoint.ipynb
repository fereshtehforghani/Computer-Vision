## Seam Carving Implementation (CarvingHelper and MySeamCarving Functions)

```python
def CarvingHelper(img, mode, seams_to_carve):
    # tensor
    carved_img = torch.tensor(img, dtype=torch.float32)
    carved_img = torch.unsqueeze(carved_img, dim=0)
    carved_img = torch.permute(carved_img, (0, 3, 1, 2))
    for i in range(seams_to_carve):
        if i%50 == 0:
            print(str(i) + " seams removed")

        #gray scale
        tensor_img_gray = kornia.color.rgb_to_grayscale(carved_img)
        grads = kornia.filters.SpatialGradient()(tensor_img_gray)[0, 0]

        #energy computations
        energy = torch.sqrt(torch.pow(grads[0, :, :], 2) + torch.pow(grads[1, :, :], 2))

        # Compute M with DP
        M = torch.zeros((carved_img.shape[2], carved_img.shape[3]))
        M[0, :] = energy[0, :]
        for row in range(1, M.shape[0]):
            for col in range(0, M.shape[1]):
                if col == 0:
                    M[row, col] = energy[row, col] + min(M[row-1, col: col+2])
                elif col == M.shape[1] - 1:
                    M[row, col] = energy[row, col] + min(M[row-1, col-1:])
                else:
                    M[row, col] = energy[row, col] + min(M[row-1, col-1: col+2])
        
        # find and remove min seam
        img_out = carved_img[:, :, :, :M.shape[1]-1].clone()
        rows = img_out.shape[2]
        last_row_min = torch.argmin(M[-1, :]).item()
        for i in range(rows-1, -1, -1):
            img_out[:, :, i, last_row_min:] = carved_img[:, :, i, last_row_min+1 :]
            if i == 0:
                break
            if last_row_min == 0:
                last_row_min = torch.argmin(M[i-1, last_row_min: last_row_min + 2]).item() + last_row_min
            elif last_row_min == col - 1:
                last_row_min = torch.argmin(M[i-1, last_row_min - 1: last_row_min + 1]).item() + last_row_min - 1
            else:
                last_row_min = torch.argmin(M[i-1, last_row_min - 1: last_row_min + 2]).item() + last_row_min - 1
        carved_img = img_out.clone()
    img_out = torch.permute(img_out[0], (1, 2, 0))
    return img_out
```

```python
def MySeamCarving(img, img_out_shape, mode="vertical"):    
    seams_to_carve = 0
    if mode == "vertical":
        seams_to_carve = img.shape[1] - img_out_shape[1]
        seam_carved_img = CarvingHelper(img.copy(), mode, seams_to_carve)
    elif mode == "horizontal":
        seams_to_carve = img.shape[0] - img_out_shape[0]
        seam_carved_img = CarvingHelper(np.transpose(img.copy(), (1, 0, 2)), mode, seams_to_carve)
    return seam_carved_img
```

## Seam Enlarging Implementation (EnlargingHelper and MySeamEnlarging Functions)

```python
def EnlargingHelper(img, mode, seams_to_enlarge):
    # tensor
    enlaged_img = torch.tensor(img / 255, dtype=torch.float32)
    enlaged_img = torch.unsqueeze(enlaged_img, dim=0)
    enlaged_img = torch.permute(enlaged_img, (0, 3, 1, 2))
    enlaged_img_copy = enlaged_img.clone()
    min_seams = []
    for i in range(seams_to_enlarge):
        print(str(i) + " min seams found")
        #gray scale
        tensor_img_gray = kornia.color.rgb_to_grayscale(enlaged_img)
        grads = kornia.filters.SpatialGradient()(tensor_img_gray)[0, 0]

        #energy computations
        energy = torch.sqrt(torch.pow(grads[0, :, :], 2) + torch.pow(grads[1, :, :], 2))

        # Compute M with DP
        M = torch.zeros((enlaged_img.shape[2], enlaged_img.shape[3]))
        M[0, :] = energy[0, :]
        for row in range(1, M.shape[0]):
            for col in range(0, M.shape[1]):
                if col == 0:
                    M[row, col] = energy[row, col] + min(M[row-1, col: col+2])
                elif col == M.shape[1] - 1:
                    M[row, col] = energy[row, col] + min(M[row-1, col-1:])
                else:
                    M[row, col] = energy[row, col] + min(M[row-1, col-1: col+2])
        
        # find min seams
        img_out = enlaged_img[:, :, :, :M.shape[1]-1].clone()
        rows = img_out.shape[2]
        last_row_min = torch.argmin(M[-1, :]).item()
        min_seam = [last_row_min]
        for i in range(rows-1, -1, -1):
            img_out[:, :, i, last_row_min:] = enlaged_img[:, :, i, last_row_min+1 :]
            if i == 0:
                break
            if last_row_min == 0:
                last_row_min = torch.argmin(M[i-1, last_row_min: last_row_min + 2]).item() + last_row_min
            elif last_row_min == col - 1:
                last_row_min = torch.argmin(M[i-1, last_row_min - 1: last_row_min + 1]).item() + last_row_min - 1
            else:
                last_row_min = torch.argmin(M[i-1, last_row_min - 1: last_row_min + 2]).item() + last_row_min - 1
            min_seam.append(last_row_min)
        enlaged_img = img_out.clone()
        min_seams.append(min_seam)
    
    for i in range(seams_to_enlarge):
        if i%50 == 0:
            print(str(i) + " seams added")
        #add min seams
        img_out = torch.zeros((1, 3, enlaged_img_copy.shape[2], enlaged_img_copy.shape[3]+1))
        for j in range(len(min_seams[i])):
            s = min_seams[i][j]
            img_out[0, :, j, :s + 1] = enlaged_img_copy[0, :, j, :s + 1]
            for channel in range(3):
                img_out[0, channel, j, s + 1] = torch.mean(enlaged_img_copy[0, channel, j, s-1: s + 2])
            img_out[0, :, j, s + 2:] = enlaged_img_copy[0, :, j, s + 1:]
        enlaged_img_copy = img_out.clone()
    img_out = torch.permute(img_out[0], (1, 2, 0))
    return img_out
```

```python
def MySeamEnlarging(img, img_out_shape, mode="vertical"):    
    seams_to_carve = 0
    if mode == "vertical":
        seams_to_add = img_out_shape[1] - img.shape[1]
        seam_enlarged_img = EnlargingHelper(img.copy(), mode, seams_to_add)
    elif mode == "horizontal":
        seams_to_add = img.shape[0] - img_out_shape[0]
        seam_enlarged_img = EnlargingHelper(np.transpose(img.copy(), (1, 0, 2)), mode, seams_to_add)
    return seam_enlarged_img
```

```python
img_out_shape = (1151, 1800)
mode = "vertical"
seam_enlarged_img = MySeamEnlarging(img, img_out_shape, mode)
print(seam_enlarged_img.shape)
plt.imshow(seam_enlarged_img)
```

```python
save_img = Image.fromarray(np.uint8(seam_enlarged_img))
save_img = save_img.save('York_Enlarged(1151, 1800).jpg')
```
